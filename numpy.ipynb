Initializing Different Types of Arrays

```python
# All 0s Matrix
np.zeros([2,3]) 
```

```python
# All 1s Matrix
np.ones((2,3), dtype='int32')
```

```python
# Any other number shape,number
np.full((2,2), 99)
```

```python
# Any other number (Full_Like)
np.full_like(c,4)
```

```python
# Random decimal numbers
# np.random.rand(2,4)
np.random.random_sample(c.shape)
```

```python
# Random Integer values
np.random.randint(7, size=(3,3))
```

```python
# Identity Matrix
np.identity(5)
```

```python
# Repeat an Array
arr = np.array([[1,2,3]])
r1 = np.repeat(arr, 3, axis=0)
print(r1)
```

```python
output = np.ones((5,5))
print(output)

z = np.zeros((3,3))
print(z)
z[1,1] = 9
print(z)

output[1:4, 1:4] = z
print(output)
```

Be careful when copying arrays
